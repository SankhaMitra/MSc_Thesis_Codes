```python
## These are the python codes for replication of the results given in my M.Sc. Thesis
## Kindly note that the results are preliminary, and therefore, not to be cited
import pandas as pd 
data_path = r'D:\Semester4_Courses\HSP620\Results_Replication_Python\PanelData_1991to2019_Varsinpercent.csv'
## Kindly change the data path in accordance with your working directory
data_replication = pd.read_csv(data_path) 
```

```python
data_replication
```

```python
data_replication['sfe'] = pd.Categorical(data_replication['statecode'])
data_replication['tfe'] = pd.Categorical(data_replication['Year'])
data_replication.set_index(['statecode', 'Year'])
```

```python
import warnings 
warnings.filterwarnings('ignore')
warnings.warn('DelftStack')
warnings.warn('Do not show this message') 
## This is to stop displaying the warning messages which are visually unaesthetic
## One may however wish to display the warning messages 
## Remove the codes from "import warnings" to "warnings.warn('Do not show this message')" if you want to display the warning messages 
import statsmodels.formula.api as sm 
## Standard errors clustered at state-level are used
model_1 = sm.ols(formula = 'schools_established ~ SC_res + ST_res + SC_curr_share + ST_curr_share + sc_census_share + st_census_share + SC_res_inter_SC_curr_share + ST_res_inter_ST_curr_share + female_win_prop + Elec_Yrs_ECI + Population_Density_SqKm + C(statecode) + C(Year) + Year*C(statecode)', data = data_replication).fit(cov_type = 'cluster', cov_kwds = {'groups': data_replication['statecode']})
model_1.summary()
```

```python
model_2 = sm.ols(formula = 'sc_schools ~ SC_res + ST_res + SC_curr_share + ST_curr_share + sc_census_share + st_census_share + SC_res_inter_SC_curr_share + ST_res_inter_ST_curr_share + female_win_prop + Elec_Yrs_ECI + Population_Density_SqKm + C(statecode) + C(Year) + Year*C(statecode)', data = data_replication).fit(cov_type = 'cluster', cov_kwds = {'groups': data_replication['statecode']})
model_2.summary()
```

```python
model_3 = sm.ols(formula = 'st_schools ~ SC_res + ST_res + SC_curr_share + ST_curr_share + sc_census_share + st_census_share + SC_res_inter_SC_curr_share + ST_res_inter_ST_curr_share + female_win_prop + Elec_Yrs_ECI + Population_Density_SqKm + C(statecode) + C(Year) + Year*C(statecode)', data = data_replication).fit(cov_type = 'cluster', cov_kwds = {'groups': data_replication['statecode']})
model_3.summary()
```

```python
model_4 = sm.ols(formula = 'tot_min_schools ~ SC_res + ST_res + SC_curr_share + ST_curr_share + sc_census_share + st_census_share + SC_res_inter_SC_curr_share + ST_res_inter_ST_curr_share + female_win_prop + Elec_Yrs_ECI + Population_Density_SqKm + C(statecode) + C(Year) + Year*C(statecode)', data = data_replication).fit(cov_type = 'cluster', cov_kwds = {'groups': data_replication['statecode']})
model_4.summary()
```

```python
## I present all the three models in 1 table, I do not report the fixed effects, additional controls in the table 
from statsmodels.iolib.summary2 import summary_col 
replication_results_table = summary_col(results = [model_1, model_2, model_3, model_4], stars  = True, float_format = '%0.2f', model_names = ['schools_state', 'sc_dist_schools', 'st_dist_schools', 'minority_dist_schools'], info_dict = {'N': lambda x: "{0:d}".format(int(x.nobs))} , regressor_order = ['SC_res', 'ST_res', 'SC_curr_share', 'ST_curr_share', 'SC_res_inter_SC_curr_share', 'ST_res_inter_ST_curr_share'], drop_omitted = True)
replication_results_table.add_title('Table-Regression Results (M.Sc. Thesis)')
print(replication_results_table)
```
